```r

# Load necessary libraries
library(conflicted)
library(tidyverse)
```

```r
# Load the dataset
data(iris)

# View the first few rows
iris %>%
  slice_head(n = 6) 

```

```r
iris %>%
  summary()
```

```r
plot(iris$Petal.Length, iris$Petal.Width)
```

```r

# Fit a linear model
model <- lm(Petal.Width ~ Petal.Length, data=iris)

# Plot the data and add the regression line
plot(iris$Petal.Length, iris$Petal.Width, main="Petal Width vs Petal Length",
     xlab="Petal Length", ylab="Petal Width", pch=19, col="blue")
abline(model, col="red")

```

```r

# Diagnostic plots
par(mfrow=c(2,2)) # Arrange plots in a 2x2 grid
plot(model)
```

```r
data(mtcars)

# View the first few rows
mtcars %>%
  slice_head(n = 6) %>%
  print()
```

```r
mtcars %>%
  summary()
```

```r
mycars = select( mtcars, mpg, cyl, hp, wt )

# Use the pairs function to create a matrix of scatterplots
pairs(mycars, main="Cars data - Pairs Plot", pch=21)


# Load the GGally package
library(GGally)

# Use the ggpairs function to create a matrix of scatterplots with ggplot2
ggpairs(mycars,title="Cars Data - Pairs Plot")

```

```r

# Fit the model
model <- lm(mpg ~ cyl + hp + wt, data = mtcars) 

# Display model summary

summary(model)



```

```r

# Diagnostic plots
par(mfrow=c(2,2)) # Arrange plots in a 2x2 grid
plot(model)
```

```r

# Load the necessary library
library(car)

# Calculate VIF
vif_result <- vif(model)

# Print the VIF result
print(vif_result)

# Interpretation
```
